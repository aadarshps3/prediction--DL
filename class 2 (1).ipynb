```python
import pandas as pd
import numpy as np
import sklearn
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
```

```python
X=digits.data
y=digits.target
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,train_size=.75,random_state=1)
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(xtrain,ytrain)
model.score(xtest,ytest)
```

```python
from sklearn.linear_model import LogisticRegression
model1=LogisticRegression()
model1.fit(xtrain,ytrain)
model1.score(xtest,ytest)
```

```python
yact=ytest
ypred=model1.predict(xtest)
from sklearn.metrics import confusion_matrix
mat=confusion_matrix(yact,ypred)
from mlxtend.plotting import plot_confusion_matrix
plot_confusion_matrix(mat)
```

```python
from sklearn.tree import DecisionTreeClassifier
model_1=DecisionTreeClassifier()
model_1.fit(xtrain,ytrain)
model_1.score(xtest,ytest)
```

```python
yact=ytest
ypred=model_1.predict(xtest)
from sklearn.metrics import confusion_matrix
mat=confusion_matrix(yact,ypred)
from mlxtend.plotting import plot_confusion_matrix
plot_confusion_matrix(mat)
```

```python
from sklearn.datasets import load_digits
```

```python
digits=load_digits()
X=digits['data']
y=digits['target']
```

```python
from sklearn.model_selection import KFold
```

```python
fld=KFold(5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

# working of the cv

```python
log_score=[]
dt_score=[]
rf_score=[]
```

```python
for i in fld.split(X,y):
    train_index,test_index=i
    xtrain=X[train_index]
    ytrain=y[train_index]
    xtest=X[test_index]
    ytest=y[test_index]
    #logistic model
    lmodel=LogisticRegression()
    lmodel.fit(xtrain,ytrain)
    log_score.append(lmodel.score(xtest,ytest))
    # decision tree model
    dt_model=DecisionTreeClassifier()
    dt_model.fit(xtrain,ytrain)
    dt_score.append(dt_model.score(xtest,ytest))
    #Random forest model
    rf_model=RandomForestClassifier()
    rf_model.fit(xtrain,ytrain)
    rf_score.append(rf_model.score(xtest,ytest))
    print('______________________________________________________')
    
```

```python
log_score
```

```python
np.array(log_score).mean()
```

```python
np.array(dt_score).mean()
```

```python
np.array(rf_score).mean()
```
